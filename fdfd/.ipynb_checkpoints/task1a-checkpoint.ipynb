```python
%matplotlib inline
from matplotlib import pyplot as plt
    
import numpy as np
import scipy.sparse.linalg as ssl
    
#    n = 20# number of steps
nlist = [6, 10,30,50,100]
    
for n in nlist:
    c = 1 #spd of light
    size = 1   # domain size
    
    dx = size/n  #step size
    getev = 4 # eigenvalues desired
    x = np.arange(0,size,dx)
    
    iden = np.ones((n,n)) #identity matrix
    F = (1/dx)*(np.triu(iden)-2*np.diag(np.diag(iden))-np.triu(iden,2))
    B = (1/dx)*(-np.tril(iden)+2*np.diag(np.diag(iden))+np.tril(iden,-2))
    bf = -np.matmul(B,F)
    
    #%matplotlib inline
    #import matplotlib.pyplot as plt
    #import scipy as sp 
    #plt.spy(sp.sparse.csr_matrix(bf)) # to visualize matrix
    #plt.show()
    
    evalues, evectors = ssl.eigs(bf, k=getev, M=None, sigma=-(2*np.pi)**2, v0 = np.ones(n)) 

#    k = np.sort(np.sqrt(evalues))
    
    #print(np.linalg.eigvals(bf-kt**2))
    #print(np.real(k))
    
#    lam = 2*np.pi*c/np.real(np.sort(np.sqrt(evalues))) # this is actually angularfreq
    lam = 1/np.real(np.sort(np.sqrt(evalues)))
    #qualf = np.real(k)/(2*np.imag(np.sqrt(evalues)))) #is infinite?
    
    plt.hold(True)
    fignum = 0
    for i in range(1, getev):
        fignum = fignum + 1
        plt.figure(fignum)
        plt.plot(x, np.sqrt(n)*np.real(evectors[:,i]), label="n = " + str(n))
        plt.legend()
        
        
    plt.figure(fignum+1)
    plt.xlabel("eigenvalue")
    plt.ylabel("$\lambda$")
    plt.title("$\lambda$ vs eigenvalue")
    plt.plot(lam, label="n = " + str(n))
    plt.legend()
```
